### üåü 1Cycle

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
         .format(MSE, RMSE, R2))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = enc_r_df.iloc[:, :-1], enc_r_df.iloc[:, -1]

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏
l_r = LinearRegression()
# ÌõàÎ†®
l_r.fit(X_train, y_train)
```

```python
# ÏòàÏ∏°
prediction = l_r.predict(X_test)
# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```

#### üí° Î∂ÑÏÑù
- LinearRegressionÏùÑ ÌÜµÌïú ÌïôÏäµ Í≤∞Í≥º R2 Ï†êÏàòÍ∞Ä 1.0Î°ú Î≥¥Ïù¥Í≥† ÏûàÏùå
- Ï∂îÍ∞ÄÏ†ÅÏù∏ Ï†ÑÏ≤òÎ¶¨ Í≥ºÏ†ïÏùÑ ÌÜµÌï¥ ÏàòÏπòÎ•º Ï°∞Ï†ïÌï¥Î≥¥Î†§ Ìï®

```python
# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Ïù¥ÏÉÅÏπò Ï†úÍ±∞
from sklearn.preprocessing import StandardScaler

dro_r_df = enc_r_df.copy()

# ÌëúÏ§ÄÌôî Í∞ùÏ≤¥ ÏÉùÏÑ±
std = StandardScaler()
# Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî Î≥ÄÌôò
result = std.fit_transform(dro_r_df)
# ÌëúÏ§ÄÌôî Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
std_r_df = pd.DataFrame(result, columns=dro_r_df.columns)

for column in std_r_df.columns:
    # ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Ïù¥ÏÉÅÏπò Ï†úÍ±∞
    copy_std_r_df = std_r_df[std_r_df[column].between(-1.96, 1.96)]

# Ïù¥ÏÉÅÏπò Ï†úÍ±∞ÌõÑ Ïú†ÏßÄÎêòÎäî Ïù∏Îç±Ïä§Î•º ÎåÄÏ°∞ÌïòÏó¨ ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ Ïù¥ÏÉÅÏπò Ï†úÍ±∞
dro_r_df = dro_r_df.iloc[copy_std_r_df.index].reset_index(drop=True)
dro_r_df
```

### üåü 3Cycle
Ï∞®Ïõê Ï∂ïÏÜå

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = dro_r_df.iloc[:, :-1], dro_r_df.iloc[:, -1]

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ÌïôÏäµ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î°ú Î≥ëÌï©
train_df = pd.concat([X_train, y_train], axis=1).reset_index(drop=True)
test_df = pd.concat([X_test, y_test], axis=1).reset_index(drop=True)
```

```python
from sklearn.decomposition import PCA

# Ï∞®Ïõê Ï∂ïÏÜå Í∞ùÏ≤¥ ÏÉùÏÑ±
# n_components: ÏñºÎßàÎÇò Ï∞®ÏõêÏ∂ïÏÜåÌï† Í±¥ÏßÄ ÏûëÏÑ±
pca = PCA(n_components=5)

# ÌÉÄÍ≤üÏùÑ Ï†úÏô∏ÌïòÍ≥† ÌîºÏ≤òÎßå Ï∞®Ïõê Ï∂ïÏÜå
pca_train = pca.fit_transform(train_df.iloc[:, :-1])
pca_test = pca.fit_transform(test_df.iloc[:, :-1])

# ÌÅ¨Í∏∞ ÌôïÏù∏
print(pca_train.shape, pca_test.shape)
```

```python
# train
# Ïª¨ÎüºÎ™Ö ÏÑ§Ï†ï
pca_columns = [f'pca{i + 1}' for i in range(pca_train.shape[1])]

# Ï∞®Ïõê Ï∂ïÏÜå Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
pca_train_df = pd.DataFrame(pca_train, columns=pca_columns)
# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Ï∂îÍ∞Ä
pca_train_df.loc[:, 'target'] = train_df['target']

# test
# Ïª¨ÎüºÎ™Ö ÏÑ§Ï†ï
pca_columns = [f'pca{i + 1}' for i in range(pca_test.shape[1])]

# Ï∞®Ïõê Ï∂ïÏÜå Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
pca_test_df = pd.DataFrame(pca_test, columns=pca_columns)
# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Ï∂îÍ∞Ä
pca_test_df.loc[:, 'target'] = test_df['target']
```

```python
# Î≥¥Ï°¥Ïú® ÌôïÏù∏
# Í∞ÅÍ∞ÅÏùò ÌîºÏ≤òÍ∞Ä ÏñºÎßåÌÅº ÏÑ§Î™ÖÌïòÍ≥† ÏûàÎäîÏßÄ
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```

```python
import seaborn as sns

sns.pairplot(pca_train_df, hue='target')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = dro_r_df.iloc[:, :-1], dro_r_df.iloc[:, -1]

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏ Í∞ùÏ≤¥ ÏÉùÏÑ±
l_r = LinearRegression()

# ÌååÏù¥ÌîÑ ÎùºÏù∏ Íµ¨Ï∂ï
pipe = Pipeline([('pca', PCA(n_components=5)), ('l_r', l_r)])

# ÌõàÎ†®
pipe.fit(X_train, y_train)
```

```python
# ÏòàÏ∏°
prediction = pipe.predict(X_test)

# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```

#### üìä ÏÑ†Ìòï ÌöåÍ∑Ä + StandardScaler

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = dro_r_df.iloc[:, :-1], dro_r_df.iloc[:, -1]

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏ Í∞ùÏ≤¥ ÏÉùÏÑ±
l_r = LinearRegression()

# ÌååÏù¥ÌîÑ ÎùºÏù∏ Íµ¨Ï∂ï
pipe = Pipeline([('std', StandardScaler()), ('pca', PCA(n_components=5)), ('l_r', l_r)])

# ÌõàÎ†®
pipe.fit(X_train, y_train)
```

```python
# ÏòàÏ∏°
prediction = pipe.predict(X_test)

# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
from sklearn.preprocessing import PolynomialFeatures

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = dro_r_df.iloc[:, :-1], dro_r_df.iloc[:, -1]

# Ï∞®Ïàò ÌôïÏû•
poly_features = PolynomialFeatures(degree=2).fit_transform(features)

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏ Í∞ùÏ≤¥ ÏÉùÏÑ±
l_r = LinearRegression()

# ÌååÏù¥ÌîÑ ÎùºÏù∏ Íµ¨Ï∂ï
pipe = Pipeline([('pca', PCA(n_components=5)), ('l_r', l_r)])

# ÌõàÎ†®
pipe.fit(X_train, y_train)
```

```python
# ÏòàÏ∏°
prediction = pipe.predict(X_test)
# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = dro_r_df.iloc[:, :-1], dro_r_df.iloc[:, -1]

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ÌöåÍ∑Ä Î™®Îç∏ Îã¥Í∏∞
dt_r = DecisionTreeRegressor(random_state=124)
rf_r = RandomForestRegressor(random_state=124, n_estimators=1000)
gb_r = GradientBoostingRegressor(random_state=124)
xgb_r = XGBRegressor(random_state=124)
lgb_r = LGBMRegressor(random_state=124)

models = [dt_r, rf_r, gb_r, xgb_r, lgb_r]

# Î™®Îç∏ Î≥Ñ ÌïôÏäµ ÏßÑÌñâ:
for model in models:
    # ÌååÏù¥ÌîÑ ÎùºÏù∏ Íµ¨Ï∂ï
    pipe = Pipeline([('pca', PCA(n_components=5)), (f'{model}', model)])
    # Î™®Îç∏ ÌïôÏäµ
    pipe.fit(X_train, y_train)
    # ÏòàÏ∏°
    prediction = pipe.predict(X_test)
    # Î™®Îç∏Î™Ö Ï∂úÎ†• (Î™®Îç∏Î™ÖÍ≥º ÌèâÍ∞Ä Í≤∞Í≥ºÍ∞Ä Í∞ôÏù¥ Ï∂úÎ†•ÎêòÎèÑÎ°ù)
    print(model.__class__.__name__)
    # ÌèâÍ∞Ä
    get_evaluation(y_test, prediction)
```

- ÌöåÍ∑Ä Î™®Îç∏ Ï§ë GradientBoostingRegressorÏùÑ ÏÇ¨Ïö©ÌïòÏòÄÏùÑ Îïå, Í∞ÄÏû• Ï¢ãÏùÄ ÏÑ±Îä•ÏùÑ Î≥¥ÏûÑÏóê Îî∞Îùº Ìï¥Îãπ Î™®Îç∏ÏùÑ ÏÑ†ÌÉùÌïòÎäî Í≤ÉÏù¥ Ï†ÅÌï©ÌïòÎã§Í≥† ÌåêÎã®Îê®

### üåü 4Cycle
GridSearchCVÎ•º ÌÜµÌï¥ ÏµúÏ†ÅÏùò ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ Ï°∞Ìï©ÏùÑ Ï∞æÍ≥† ÏÑ±Îä•ÏùÑ Îçî Ìñ•ÏÉÅÏãúÌÇ¨ Ïàò ÏûàÎäîÏßÄ ÌôïÏù∏

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.decomposition import PCA

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = dro_r_df.iloc[:, :-1], dro_r_df.iloc[:, -1]

# Î¨∏Ï†ú/Ï†ïÎãµ Î∞è ÌïôÏäµ/ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# Í≤ÄÏ¶ù Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
val_X_train, val_X_test, val_y_train, val_y_test = \
train_test_split(X_train, y_train, test_size=0.2, random_state=124)

# Í∞ÄÏû• ÏÑ±Îä•Ïù¥ Ïö∞ÏàòÌïú Î™®Îç∏ Îã¥Í∏∞
kf_gb_r = GradientBoostingRegressor(random_state=124)

# ÌååÎùºÎØ∏ÌÑ∞ Í∞í Ï°∞Ï†ï
parameters = {'kf_gb_r__max_depth': [6, 7, 8], 'kf_gb_r__min_samples_split': [60, 70, 80], 'kf_gb_r__n_estimators': [60, 70, 80]}

# ÍµêÏ∞®Í≤ÄÏ¶ù
# n_splits: Îç∞Ïù¥ÌÑ∞Î•º Î™á Í∞úÏùò Ìè¥ÎìúÎ°ú ÎÇòÎàåÏßÄÎ•º Í≤∞Ï†ï (ÏùºÎ∞òÏ†ÅÏúºÎ°ú 5 ÎòêÎäî 10)
# shuffle: Î∂ÑÌï† Ï†Ñ Îç∞Ïù¥ÌÑ∞ ÌòºÌï© Ïó¨Î∂Ä 
kfold = KFold(n_splits=5, random_state=124, shuffle=True)

pipe = Pipeline([('pca', PCA(n_components=5)), ('kf_gb_r', kf_gb_r)])

# ÌïôÏäµ Î∞è ÍµêÏ∞® Í≤ÄÏ¶ù Î™®Îç∏ ÏÑ§Ï†ï
grid_kf_gb_r = GridSearchCV(pipe, param_grid=parameters, cv=kfold, n_jobs=-1)

# ÌõàÎ†®
grid_kf_gb_r.fit(X_train, y_train)
```

```python
# ÌõàÎ†® Í≤∞Í≥º ÌôïÏù∏
result_df = pd.DataFrame(grid_kf_gb_r.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)
```

```python
# ÏµúÏ†ÅÏùò Î™®Îç∏ Îã¥Í∏∞
kf_gb_r = grid_kf_gb_r.best_estimator_

# ÏòàÏ∏°
prediction = kf_gb_r.predict(X_test)
# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```
