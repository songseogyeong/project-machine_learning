```python
import pandas as pd

# ë°ì´í„° í˜¸ì¶œ
r_df = pd.read_csv('./datasets/ResidentialBuildingEnergyConsumption.csv', low_memory=False)
r_df
```

```python
# ë°ì´í„° í”„ë ˆì„ ë³µì œ
pre_r_df = r_df.copy()
```

```python
# ê²°ì¸¡ì¹˜ ì œê±°
pre_r_df = pre_r_df[~pre_r_df.isna().any(axis=1)].reset_index(drop=True)
```

```python
# ë¶ˆí•„ìš”í•œ í”¼ì²˜ ì œê±°
pre_r_df = pre_r_df.drop(labels='DOEID', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

# ë°ì´í„° í”„ë ˆì„ ë³µì œ
enc_r_df = pre_r_df.copy()
encoders = {}

# ë¬¸ìì—´ ì»¬ëŸ¼ ì¶”ì¶œ
columns = columns = ['METROMICRO', 'UATYP10', 'CLIMATE_REGION_PUB', 'IECC_CLIMATE_PUB']

# ë°˜ë³µí•˜ì—¬ ì»¬ëŸ¼ ì¸ì½”ë”© ì²˜ë¦¬:
for column in columns:
    # ë ˆì´ë¸”ì¸ì½”ë” ê°ì²´ ìƒì„±
    encoder = LabelEncoder()
    # ë¬¸ìì—´ ë°ì´í„° ì •ìˆ˜ë¡œ í˜•ë³€í™˜
    result = encoder.fit_transform(enc_r_df[column])
    # í˜•ë³€í™˜ ê°’ìœ¼ë¡œ ëŒ€ì²´
    enc_r_df[column] = result
    # ì›ë³¸ ë°ì´í„° ë‹´ê¸°
    encoders[column] = encoder.classes_
```

```python
# íƒ€ê²Ÿ ì»¬ëŸ¼ ìœ„ì¹˜ ë³€ê²½ì„ ìœ„í•´ ìƒˆë¡œìš´ ì»¬ëŸ¼ì— ë°ì´í„° ë‹´ê¸°
enc_r_df['target'] = enc_r_df.TOTALDOL
# ê¸°ì¡´ ì»¬ëŸ¼ì€ ì œê±°
enc_r_df = enc_r_df.drop(labels='TOTALDOL', axis=1)
```

```python
# íƒ€ê²Ÿ ë°ì´í„° ì´ìƒì¹˜ ì œê±°
from sklearn.preprocessing import StandardScaler

# í‘œì¤€í™” ê°ì²´ ìƒì„±
std = StandardScaler()
# ë°ì´í„° í‘œì¤€í™” ë³€í™˜
result = std.fit_transform(enc_r_df)
# í‘œì¤€í™” ë°ì´í„° í”„ë ˆì„ ìƒì„±
std_r_df = pd.DataFrame(result, columns=enc_r_df.columns)

for column in std_r_df.columns:
    # íƒ€ê²Ÿ ë°ì´í„° ì´ìƒì¹˜ ì œê±°
    copy_std_r_df = std_r_df[std_r_df[column].between(-1.96, 1.96)]

# ì´ìƒì¹˜ ì œê±°í›„ ìœ ì§€ë˜ëŠ” ì¸ë±ìŠ¤ë¥¼ ëŒ€ì¡°í•˜ì—¬ ì›ë³¸ ë°ì´í„° ì´ìƒì¹˜ ì œê±°
enc_r_df = enc_r_df.iloc[copy_std_r_df.index].reset_index(drop=True)
```

### ğŸŒŸ 2Cycle

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
         .format(MSE, RMSE, R2))
```

#### ğŸ“Š ì°¨ì› ì¶•ì†Œ

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# ë°ì´í„° ì„¸íŠ¸ ë¶„ë¦¬
# í”¼ì²˜, íƒ€ê²Ÿ ë°ì´í„° ë¶„ë¦¬
features, targets = enc_r_df.iloc[:, :-1], enc_r_df.iloc[:, -1]

# í•™ìŠµ/í…ŒìŠ¤íŠ¸ ë° ë¬¸ì œ/ì •ë‹µ ë°ì´í„° ì„¸íŠ¸ ë¶„ë¦¬
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)
```

```python
# í•™ìŠµ ë° í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ë³‘í•©
train_df = pd.concat([X_train, y_train], axis=1).reset_index(drop=True)
test_df = pd.concat([X_test, y_test], axis=1).reset_index(drop=True)
```

```python
from sklearn.decomposition import PCA

# ì°¨ì› ì¶•ì†Œ ê°ì²´ ìƒì„±
# n_components: ì–¼ë§ˆë‚˜ ì°¨ì›ì¶•ì†Œí•  ê±´ì§€ ì‘ì„±
pca = PCA(n_components=5)

# íƒ€ê²Ÿì„ ì œì™¸í•˜ê³  í”¼ì²˜ë§Œ ì°¨ì› ì¶•ì†Œ
pca_train = pca.fit_transform(train_df.iloc[:, :-1])
pca_test = pca.fit_transform(test_df.iloc[:, :-1])

# í¬ê¸° í™•ì¸
print(pca_train.shape, pca_test.shape)
```

```python
# train
# ì»¬ëŸ¼ëª… ì„¤ì •
pca_columns = [f'pca{i + 1}' for i in range(pca_train.shape[1])]

# ì°¨ì› ì¶•ì†Œ ë°ì´í„° í”„ë ˆì„ ìƒì„±
pca_train_df = pd.DataFrame(pca_train, columns=pca_columns)
# íƒ€ê²Ÿ ë°ì´í„° ì¶”ê°€
pca_train_df.loc[:, 'target'] = train_df['target']

# test
# ì»¬ëŸ¼ëª… ì„¤ì •
pca_columns = [f'pca{i + 1}' for i in range(pca_test.shape[1])]

# ì°¨ì› ì¶•ì†Œ ë°ì´í„° í”„ë ˆì„ ìƒì„±
pca_test_df = pd.DataFrame(pca_test, columns=pca_columns)
# íƒ€ê²Ÿ ë°ì´í„° ì¶”ê°€
pca_test_df.loc[:, 'target'] = test_df['target']
```

### ğŸŒŸ 4Cycle
StandardScaler, MinMaxScalerë¥¼ í†µí•œ ì „ì²˜ë¦¬

#### ğŸ“Š StandardScaler

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 

# ë°ì´í„° ì„¸íŠ¸ ë¶„ë¦¬
# í”¼ì²˜, íƒ€ê²Ÿ ë°ì´í„° ë¶„ë¦¬
features, targets = enc_r_df.iloc[:, :-1], enc_r_df.iloc[:, -1]

# ë¬¸ì œ/ì •ë‹µ ë° í•™ìŠµ/í›ˆë ¨ ë°ì´í„° ë¶„ë¦¬
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# íŒŒì´í”„ ë¼ì¸ êµ¬ì¶•
pipe = Pipeline([('std', StandardScaler()), ('pca', PCA(n_components=5)), ('gb_r', GradientBoostingRegressor(random_state=124))])

# íŒŒë¼ë¯¸í„° ê°’ ì¡°ì •
parameters = {
    'gb_r__n_estimators': [55, 60, 65],
    'gb_r__learning_rate': [0.1],
    'gb_r__max_depth': [4],
    'gb_r__min_samples_split': [39, 40, 41]
}

# êµì°¨ê²€ì¦
kfold = KFold(n_splits=5, random_state=124, shuffle=True)

# í•™ìŠµ ë° êµì°¨ ê²€ì¦ ëª¨ë¸ ì„¤ì •
std_gb_r = GridSearchCV(pipe, param_grid=parameters, cv=kfold, n_jobs=-1)

# í›ˆë ¨
std_gb_r.fit(X_train, y_train)
```

```python
# ìµœì ì˜ ëª¨ë¸
std_gb_r = std_gb_r.best_estimator_

# ì˜ˆì¸¡
prediction = std_gb_r.predict(X_test)

# í‰ê°€
get_evaluation(y_test, prediction)
```

#### ğŸ“Š MinMaxScaler

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA 

# ë°ì´í„° ì„¸íŠ¸ ë¶„ë¦¬
# í”¼ì²˜, íƒ€ê²Ÿ ë°ì´í„° ë¶„ë¦¬
features, targets = enc_r_df.iloc[:, :-1], enc_r_df.iloc[:, -1]

# ë¬¸ì œ/ì •ë‹µ ë° í•™ìŠµ/í›ˆë ¨ ë°ì´í„° ë¶„ë¦¬
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# íŒŒì´í”„ ë¼ì¸ êµ¬ì¶•
pipe = Pipeline([('min', MinMaxScaler()), ('pca', PCA(n_components=5)), ('gb_r', GradientBoostingRegressor(random_state=124))])

# íŒŒë¼ë¯¸í„° ê°’ ì¡°ì •
parameters = {
    'gb_r__n_estimators': [55, 60, 65],
    'gb_r__learning_rate': [0.1],
    'gb_r__max_depth': [4],
    'gb_r__min_samples_split': [39, 40, 41]
}

# êµì°¨ê²€ì¦
kfold = KFold(n_splits=5, random_state=124, shuffle=True)

# í•™ìŠµ ë° êµì°¨ ê²€ì¦ ëª¨ë¸ ì„¤ì •
min_gb_r = GridSearchCV(pipe, param_grid=parameters, cv=kfold, n_jobs=-1)

# í›ˆë ¨
min_gb_r.fit(X_train, y_train)
```

```python
# ìµœì ì˜ ëª¨ë¸
min_gb_r = min_gb_r.best_estimator_

# ì˜ˆì¸¡
prediction = min_gb_r.predict(X_test)

# í‰ê°€
get_evaluation(y_test, prediction)
```
