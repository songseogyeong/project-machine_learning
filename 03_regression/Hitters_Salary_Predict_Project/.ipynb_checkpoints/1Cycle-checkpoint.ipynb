#### üìä Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨

```python
# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Ïù¥ÏÉÅÏπò Ï†úÍ±∞
from sklearn.preprocessing import StandardScaler

# ÌëúÏ§ÄÌôî Í∞ùÏ≤¥ ÏÉùÏÑ±
std = StandardScaler()
# Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî Ï†ºÌôò
result = std.fit_transform(h_df[['Salary']])
# ÌëúÏ§ÄÌôî Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
std_h_df = pd.DataFrame(result, columns=['Salary'])

# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Ïù¥ÏÉÅÏπò Ï†úÍ±∞
std_h_df = std_h_df[std_h_df.Salary.between(-1.96, 1.96)]

# ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóêÎèÑ ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Ïù¥ÏÉÅÏπò Ï†úÍ±∞ Ï†ÅÏö© ÌõÑ ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóê Îã¥Í∏∞
pre_h_df = h_df.iloc[std_h_df.index].reset_index(drop=True)

pre_h_df
```

```python
# Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨ ÌôïÏù∏
pre_h_df.hist(figsize=(12, 15), color=cmap(np.array([1])), edgecolor='black')
```

```python
# ÌÉÄÍ≤ü Ïª¨Îüº Î∂ÑÌè¨ Î≥ÄÌôò
from sklearn.preprocessing import PowerTransformer

ptf = PowerTransformer(standardize=False)
pre_h_df['Salary'] = ptf.fit_transform(pre_h_df[['Salary']])
pre_h_df
```

```python
# ÌÉÄÍ≤ü Ïª¨Îüº Î∂ÑÌè¨ Î≥ÄÌôò ÌôïÏù∏
pre_h_df.Salary.hist(color=cmap(np.array([1])), edgecolor='black')
```

```python
from sklearn.preprocessing import LabelEncoder

# Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ Î≥µÏ†ú
enc_h_df = pre_h_df.copy()
encoders = {}

# Î¨∏ÏûêÏó¥ Ïª¨Îüº Ï∂îÏ∂ú
columns = ['League', 'Division', 'NewLeague']

# Î∞òÎ≥µÌïòÏó¨ Ïª¨Îüº Ïù∏ÏΩîÎî© Ï≤òÎ¶¨:
for column in columns:
    # Î†àÏù¥Î∏î Ïù∏ÏΩîÎçî Í∞ùÏ≤¥ ÏÉùÏÑ±
    encoder = LabelEncoder()
    # Î¨∏ÏûêÏó¥ Îç∞Ïù¥ÌÑ∞ Ï†ïÏàòÎ°ú ÌòïÎ≥ÄÌôò
    result = encoder.fit_transform(enc_h_df[column])
    # ÌòïÎ≥ÄÌôò Í∞íÏúºÎ°ú ÎåÄÏ≤¥
    enc_h_df[column] = result
    # ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ Îã¥Í∏∞
    encoders[column] = encoder.classes_

# ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏
print(encoders)
```

```python
# ÌÉÄÍ≤ü Ïª¨Îüº ÏúÑÏπò Î≥ÄÍ≤ΩÏùÑ ÏúÑÌï¥ ÏÉàÎ°úÏö¥ Ïª¨ÎüºÏóê Îç∞Ïù¥ÌÑ∞ Îã¥Í∏∞
enc_h_df['target'] = enc_h_df.Salary
# Í∏∞Ï°¥ Ïª¨ÎüºÏùÄ Ï†úÍ±∞
enc_h_df = enc_h_df.drop(labels='Salary', axis=1)
enc_h_df
```

```python
# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌôïÏù∏
enc_h_df.corr()['target'].sort_values(ascending=False)[1:]
```

```python
# ÏÑ†ÏàòÏùò Í≤ΩÎ†• Îç∞Ïù¥ÌÑ∞Îì§Ïùò ÌèâÍ∑† Í∞íÏúºÎ°ú ÎåÄÏ≤¥
enc_h_df['OrtCAtBat'] = enc_h_df['CAtBat'] / enc_h_df['Years']
enc_h_df['OrtCHits'] = enc_h_df['CHits'] / enc_h_df['Years']
enc_h_df['OrtCHmRun'] = enc_h_df['CHmRun'] / enc_h_df['Years']
enc_h_df['OrtCruns'] = enc_h_df['CRuns'] / enc_h_df['Years']
enc_h_df['OrtCRBI'] = enc_h_df['CRBI'] / enc_h_df['Years']
enc_h_df['OrtCWalks'] = enc_h_df['CWalks'] / enc_h_df['Years']
```

```python
# Î∂àÌïÑÏöîÌïú ÌîºÏ≤ò, ÏÇ¨Ïö©Ïù¥ ÏôÑÎ£åÎêú ÌîºÏ≤ò Î∞è ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÎÇÆÏùÄ Ï†úÍ±∞
columns = ['Errors', 'NewLeague', 'League', 'Years']

enc_h_df = enc_h_df.drop(labels=columns, axis=1)
enc_h_df
```

#### üìä ÌöåÍ∑Ä

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_squared_log_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(MSLE)
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
         .format(MSE, RMSE, MSLE, RMSLE, R2))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = enc_h_df.iloc[:, :-1], enc_h_df.iloc[:, -1]

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏
l_r = LinearRegression()
# ÌõàÎ†®
l_r.fit(X_train, y_train)
```

```python
# ÏòàÏ∏°
prediction = l_r.predict(X_test)
# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_squared_log_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
         .format(MSE, RMSE, R2))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = enc_h_df.iloc[:, :-1], enc_h_df.iloc[:, -1]

# Ï∞®Ïõê ÌôïÏû• (1Ï∞®Ïõê)
poly_features = PolynomialFeatures(degree=2).fit_transform(features)

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(poly_features, targets, test_size=0.2, random_state=124)

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏
l_r = LinearRegression()
# ÌõàÎ†®
l_r.fit(X_train, y_train)
```

```python
# ÏòàÏ∏°
prediction = l_r.predict(X_test)
# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = enc_h_df.iloc[:, :-1], enc_h_df.iloc[:, -1]

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ÌöåÍ∑Ä Î™®Îç∏ Îã¥Í∏∞
dt_r = DecisionTreeRegressor(random_state=124)
rf_r = RandomForestRegressor(random_state=124, n_estimators=1000)
gb_r = GradientBoostingRegressor(random_state=124)
xgb_r = XGBRegressor(random_state=124)
lgb_r = LGBMRegressor(random_state=124)

models = [dt_r, rf_r, gb_r, xgb_r, lgb_r]

# Î™®Îç∏ Î≥Ñ ÌïôÏäµ ÏßÑÌñâ:
for model in models:
    # Î™®Îç∏ ÌïôÏäµ
    model.fit(X_train, y_train)
    # ÏòàÏ∏°
    prediction = model.predict(X_test)
    # Î™®Îç∏Î™Ö Ï∂úÎ†• (Î™®Îç∏Î™ÖÍ≥º ÌèâÍ∞Ä Í≤∞Í≥ºÍ∞Ä Í∞ôÏù¥ Ï∂úÎ†•ÎêòÎèÑÎ°ù)
    print(model.__class__.__name__)
    # ÌèâÍ∞Ä
    get_evaluation(y_test, prediction)
```
