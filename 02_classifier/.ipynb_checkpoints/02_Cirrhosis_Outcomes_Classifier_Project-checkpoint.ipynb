#### üìä Î¨∏ÏûêÏó¥ Îç∞Ïù¥ÌÑ∞ ÌòïÎ≥ÄÌôò

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc_c_df = d_c_df.copy()
enc_c_df
```

```python
Drug_encoder = LabelEncoder()
Drug = Drug_encoder.fit_transform(enc_c_df['Drug'].tolist())
enc_c_df['Drug'] = Drug

Sex_encoder = LabelEncoder()
Sex = Sex_encoder.fit_transform(enc_c_df['Sex'].tolist())
enc_c_df['Sex'] = Sex

Ascites_encoder = LabelEncoder()
Ascites = Ascites_encoder.fit_transform(enc_c_df['Ascites'].tolist())
enc_c_df['Ascites'] = Ascites

Hepatomegaly_encoder = LabelEncoder()
Hepatomegaly = Hepatomegaly_encoder.fit_transform(enc_c_df['Hepatomegaly'].tolist())
enc_c_df['Hepatomegaly'] = Hepatomegaly

Spiders_encoder = LabelEncoder()
Spiders = Spiders_encoder.fit_transform(enc_c_df['Spiders'].tolist())
enc_c_df['Spiders'] = Spiders

Edema_encoder = LabelEncoder()
Edema = Edema_encoder.fit_transform(enc_c_df['Edema'].tolist())
enc_c_df['Edema'] = Edema

Status_encoder = LabelEncoder()
Status = Status_encoder.fit_transform(enc_c_df['Status'].tolist())
enc_c_df['Status'] = Status
```

#### üìä Ïò§Ï∞®ÌñâÎ†¨

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction, average='macro')
    recall = recall_score(y_test , prediction, average='macro')
    f1 = f1_score(y_test, prediction, average='macro')
    
    print('Ïò§Ï∞® ÌñâÎ†¨')
    print(confusion)
    print('Ï†ïÌôïÎèÑ: {0:.4f}, Ï†ïÎ∞ÄÎèÑ: {1:.4f}, Ïû¨ÌòÑÏú®: {2:.4f}, F1: {3:.4f}'.format(accuracy, precision, recall, f1))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
prediction = dtc.predict(X_test)
```

```python
get_evaluation(y_test, prediction, dtc, X_test)
```

```python
import numpy as np

# ClassifierÏùò Decision BoundaryÎ•º ÏãúÍ∞ÅÌôî ÌïòÎäî Ìï®Ïàò
def visualize_boundary(model, X, y):
    fig,ax = plt.subplots()
    
    # ÌïôÏäµ Îç∞Ïù¥ÌÉÄ scatter plotÏúºÎ°ú ÎÇòÌÉÄÎÇ¥Í∏∞
    ax.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, s=25, cmap='rainbow', edgecolor='k',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()
    
    # Ìò∏Ï∂ú ÌååÎùºÎØ∏ÌÑ∞Î°ú Îì§Ïñ¥Ïò® training Îç∞Ïù¥ÌÉÄÎ°ú model ÌïôÏäµ . 
    model.fit(X.values, y)
    # meshgrid ÌòïÌÉúÏù∏ Î™®Îì† Ï¢åÌëúÍ∞íÏúºÎ°ú ÏòàÏ∏° ÏàòÌñâ. 
    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    # contourf() Î•º Ïù¥Ïö©ÌïòÏó¨ class boundary Î•º visualization ÏàòÌñâ. 
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow',
                           zorder=1)
```

```python
# Ï†ÑÏ≤¥ ÌñâÏùò ÌäπÏ†ï Îç∞Ïù¥ÌÑ∞ ÏÑ†ÌÉùÌïòÏó¨ ÌîºÏ≤òÎ°ú ÏßÄÏ†ï, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞Î•º ÌÉÄÍ≤üÏúºÎ°ú ÏßÄÏ†ï
features, targets = enc_c_df.iloc[:, [0, 8]], enc_c_df.iloc[:, -1]
features
```

```python
# Í≤∞Ï†ï Ìä∏Î¶¨Ïùò Î¨¥ÏûëÏúÑÏÑ±ÏùÑ Ï†úÏñ¥ÌïòÍ≥† Î∂ÑÎ•ò Î™®Îç∏ Îã¥Í∏∞
dtc = DecisionTreeClassifier(random_state=124, min_samples_leaf=70)

# Îç∞Ïù¥ÌÑ∞ ÌïôÏäµ
dtc.fit(features, targets)

# ÏãúÍ∞ÅÌôî Ìï®ÏàòÎ•º ÌÜµÌï¥ Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî
visualize_boundary(dtc, features, targets)
```

### üåü 4Cycle (ÌäπÏ†ï ÌÉÄÍ≤ü Ï†úÍ±∞, Ïñ∏Îçî ÏÉòÌîåÎßÅ)

#### üìä Ï†ÑÏ≤òÎ¶¨

```python
co_c_df['Status'].value_counts()
```

```python
co_c_df = co_c_df[co_c_df['Status'] != 1]
co_c_df['Status'].value_counts()
```

```python
# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ ÎπÑÏ§ë ÎèôÏùºÌïòÍ≤å Ï°∞Ï†ï
zero = co_c_df[co_c_df.Status == 0].sample(1736, random_state=124)
two = co_c_df[co_c_df.Status == 2]

co_c_df = pd.concat([zero, two])
co_c_df.reset_index(drop=True, inplace=True)
```

```python
co_c_df['Status'].value_counts()
```

#### üìä Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨

```python
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = co_c_df.iloc[:, :-1], co_c_df.iloc[:, -1]

# Î¨∏Ï†ú/Ï†ïÎãµ Î∞è ÌïôÏäµ/ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨ 
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, stratify=targets, test_size=0.2, random_state=124)

# Í≤ÄÏ¶ùÏö© Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_val_train, X_val_test, y_val_train, y_val_test = \
train_test_split(X_train, y_train, stratify=y_train, test_size=0.2, random_state=124)

# Í≤ÄÏ¶ùÏö© Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Îã¥Í∏∞
evals = [(X_val_train, y_val_train), (X_val_test, y_val_test)]
```

#### üìä Ï°∞Ï†ï

```python
from lightgbm import LGBMClassifier
```

```python
# ÌååÎùºÎØ∏ÌÑ∞ Í∞í ÏÑ§Ï†ï
parameters = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.3, 0.5, 0.7, 1],
    'max_depth': [10, 20, 30],
    'min_samples_split': [70, 80, 90]
}

# ÌïôÏäµ Î∞è ÍµêÏ∞® Í≤ÄÏ¶ù Î™®Îç∏ ÏÑ§Ï†ï
four_lgbm = GridSearchCV(LGBMClassifier(), param_grid=parameters, cv=3, refit=True, return_train_score=True, n_jobs=-1)
```

#### üìä ÌõàÎ†®

```python
# ÌõàÎ†®
four_lgbm.fit(X_train, y_train, eval_set=evals)
```

#### üìä ÏòàÏ∏°

```python
# ÏµúÏ†ÅÏùò Î™®Îç∏ Îã¥Í∏∞
four_lgbm = four_lgbm.best_estimator_
# ÏòàÏ∏°
prediction = four_lgbm.predict(X_test)
```

#### üìä ÌèâÍ∞Ä

```python
# Ïò§Ï∞®ÌñâÎ†¨
get_evaluation(y_test, prediction, four_lgbm, X_test)
```
