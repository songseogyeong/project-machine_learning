```python
# Ï†ÑÏ≤¥ Ïù∏Ïõê ÏÑ±ÎπÑ
gender_counts = ha_df['Gender'].value_counts().reset_index()
gender_counts['Gender'] = gender_counts['Gender'].map({1: 'male', 0: 'female'})
# Ï†ïÏÉÅ ÌôòÏûê ÏÑ±ÎπÑ
gender_ne_counts = ha_df[ha_df['Result'] == 'negative']['Gender'].value_counts().reset_index()
gender_ne_counts['Gender'] = gender_ne_counts['Gender'].map({1: 'male', 0: 'female'})
# Ïã¨Ïû• ÎßàÎπÑ ÌôòÏûê ÏÑ±ÎπÑ
gender_po_counts = ha_df[ha_df['Result'] == 'positive']['Gender'].value_counts().reset_index()
gender_po_counts['Gender'] = gender_po_counts['Gender'].map({1: 'male', 0: 'female'})
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 10))

# plt.rcParams['font.size'] = 12

# Î∂ÄÏ±ÑÍº¥ Ïä§ÌÉÄÏùº
wedgeprops = {'width': 0.7, 'edgecolor': 'w', 'linewidth': 1}

# Ïª¨Îü¨Îßµ Í∞ÄÏ†∏Ïò§Í∏∞
cmap = plt.get_cmap('Pastel1')
# Ïª¨Îü¨ÎßµÏùò ÏÉâÏÉÅ ÏÑ†ÌÉù
all_colors = cmap(np.array([0,4]))
outer_colors = cmap(np.array([1,2]))

# Í∑∏ÎûòÌîÑ Îç∞Ïù¥ÌÑ∞ Í∞ÄÏ†∏Ïò§Í∏∞
counts = gender_counts['count']
labels = gender_counts['Gender']
ne_counts = gender_ne_counts['count']
ne_labels = gender_ne_counts['Gender']
po_counts = gender_po_counts['count']
po_labels = gender_po_counts['Gender']

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
axes[0].pie(counts, labels=labels, colors=all_colors, autopct='%1.1f%%', startangle=90, wedgeprops=wedgeprops)
axes[0].set_title('Ï†ÑÏ≤¥ ÌôòÏûê ÎπÑÏú®')
axes[1].pie(ne_counts, labels=ne_labels, colors=outer_colors, autopct='%1.1f%%', startangle=90, wedgeprops=wedgeprops)
axes[1].set_title('Ï†ïÏÉÅ ÌôòÏûê ÎπÑÏú®')
axes[2].pie(po_counts, labels=po_labels, colors=outer_colors, autopct='%1.1f%%', startangle=90, wedgeprops=wedgeprops)
axes[2].set_title('Ïã¨Ïû•ÎßàÎπÑ ÌôòÏûê ÎπÑÏú®')

display(pd.concat([gender_counts.reset_index(drop=True), gender_ne_counts.reset_index(drop=True), gender_po_counts.reset_index(drop=True)], axis=1))

plt.tight_layout()
plt.show()
```

#### üìä 0ÏùÑ Í∞ÄÏßÄÎäî Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏

```python
# ÌîºÏ≤ò Í∞íÏù¥ 0Ïù∏ Îç∞Ïù¥ÌÑ∞Ïùò Í∞úÏàòÏôÄ ÎπÑÏú® ÌôïÏù∏

# ÌîºÏ≤ò zero_feauresÏóê Îã¥Í∏∞
zero_features = ['Heart rate', 'CK-MB', 'Troponin']

# Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùò Ï†ÑÏ≤¥ ÌñâÏùò ÏàòÎ•º Í∞ÄÏ†∏ÏôÄ Îã¥Í∏∞
total_count = ha_df.shape[0]

# Í∞Å zero_featuresÎì§ÏùÑ Î∞òÎ≥µÌïòÏó¨ featureÏóê ÌïòÎÇòÏî© Îã¥Í∏∞:
for feature in zero_features:
    # Í∞Å ÌîºÏ≤òÏóêÏÑú  Í∞íÏù¥ 0Ïù¥ ÎêòÎäî ÌñâÏùÑ Í∞ÄÏ†∏Ïò§Í≥† ÌñâÏùò Í∞úÏàòÎ•º ÏÑ∏Ïñ¥ Î≥ÄÏàòÏóê Îã¥Í∏∞
    zero_count = ha_df[ha_df[feature] == 0][feature].count()
    # Í∞Å ÌîºÏ≤òÎ™Ö, Í∞íÏù¥ 0Ïù∏ Îç∞Ïù¥ÌÑ∞Ïùò Í∞úÏàò, Ï†ÑÏ≤¥ ÎπÑÏú® Ï∂úÎ†•
    print(f'{feature}: {zero_count}Í±¥, {np.round(zero_count / total_count * 100, 4)}%')
```

```python
import numpy as np

# ClassifierÏùò Decision BoundaryÎ•º ÏãúÍ∞ÅÌôî ÌïòÎäî Ìï®Ïàò
def visualize_boundary(model, X, y):
    fig,ax = plt.subplots()
    
    # ÌïôÏäµ Îç∞Ïù¥ÌÉÄ scatter plotÏúºÎ°ú ÎÇòÌÉÄÎÇ¥Í∏∞
    ax.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, s=25, cmap='rainbow', edgecolor='k',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()
    
    # Ìò∏Ï∂ú ÌååÎùºÎØ∏ÌÑ∞Î°ú Îì§Ïñ¥Ïò® training Îç∞Ïù¥ÌÉÄÎ°ú model ÌïôÏäµ . 
    model.fit(X.values, y)
    # meshgrid ÌòïÌÉúÏù∏ Î™®Îì† Ï¢åÌëúÍ∞íÏúºÎ°ú ÏòàÏ∏° ÏàòÌñâ. 
    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    # contourf() Î•º Ïù¥Ïö©ÌïòÏó¨ class boundary Î•º visualization ÏàòÌñâ. 
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow',
                           zorder=1)
```

```python
features = ha_sta_df[['CK-MB', 'Troponin']] 
dtc_db = DecisionTreeClassifier(random_state=124)
visualize_boundary(dtc_db, features, targets)
```

#### üìä Ïò§Ï∞®ÌñâÎ†¨

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ÏôÄ ÏòàÏ∏° Í∞ùÏ≤¥Î•º Ï†ÑÎã¨Î∞õÎäîÎã§.
def get_evaluation_by_custom_proba(y_test, prediction, visualize=False):
#     Ïò§Ï∞® ÌñâÎ†¨
    confusion = confusion_matrix(y_test, prediction)
#     Ï†ïÌôïÎèÑ
    accuracy = accuracy_score(y_test , prediction)
#     Ï†ïÎ∞ÄÎèÑ
    precision = precision_score(y_test , prediction)
#     Ïû¨ÌòÑÏú®
    recall = recall_score(y_test , prediction)
#     F1 score
    f1 = f1_score(y_test, prediction)
#     ROC-AUC
    roc_auc = roc_auc_score(y_test, prediction)
    
    print('Ïò§Ï∞® ÌñâÎ†¨')
    print(confusion)
    print('Ï†ïÌôïÎèÑ: {0:.4f}, Ï†ïÎ∞ÄÎèÑ: {1:.4f}, Ïû¨ÌòÑÏú®: {2:.4f}, F1: {3:.4f}, ROC-AUC: {4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
    print("#" * 80)
    
    if visualize:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_predictions(y_test, prediction, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
get_evaluation_by_custom_proba(y_test, prediction, visualize=True)
```

#### üìä ROC Curve, AUC

```python
from sklearn.metrics import roc_curve

def roc_curve_plot(y_test , proba_class1):
#     ÏûÑÍ≥ÑÍ∞íÏóê Îî∞Î•∏ FPR, TPR Í∞í
    fprs, tprs, thresholds = roc_curve(y_test, proba_class1)

#     ROC CurveÎ•º plot Í≥°ÏÑ†ÏúºÎ°ú Í∑∏Î¶º. 
    plt.plot(fprs , tprs, label='ROC')
#     Í∞ÄÏö¥Îç∞ ÎåÄÍ∞ÅÏÑ† ÏßÅÏÑ†ÏùÑ Í∑∏Î¶º. 
#     TPRÍ≥º FPRÏù¥ ÎèôÏùºÌïú ÎπÑÏú®Î°ú Îñ®Ïñ¥ÏßÑÎã§Îäî Í≤ÉÏùÄ Î™®Îç∏Ïù¥ ÏñëÏÑ±Í≥º ÏùåÏÑ±ÏùÑ Íµ¨Î≥ÑÌïòÏßÄ Î™ªÌïúÎã§Îäî Í≤ÉÏùÑ ÏùòÎØ∏ÌïúÎã§.
#     Îã§Î•∏ Î∂ÑÎ•òÍ∏∞Î•º ÌåêÎã®ÌïòÍ∏∞ ÏúÑÌïú Í∏∞Ï§ÄÏÑ†ÏúºÎ°ú ÏÇ¨Ïö©ÎêòÎ©∞, 
#     ÎåÄÍ∞ÅÏÑ†Ïóê Í∞ÄÍπåÏö∏ ÏàòÎ°ù ÏòàÏ∏°Ïóê ÏÇ¨Ïö©ÌïòÍ∏∞ ÌûòÎì† Î™®Îç∏Ïù¥ÎùºÎäî ÎúªÏù¥Îã§.
    plt.plot([0, 1], [0, 1], 'k--', label='Standard')
    
    # XÏ∂ï(FPR)Ïùò ScaleÏùÑ 0.1 Îã®ÏúÑÎ°ú Î≥ÄÍ≤Ω
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1),2))
    plt.xlim(0,1); plt.ylim(0,1)
    plt.xlabel('FPR( 1 - Sensitivity )'); plt.ylabel('TPR( Recall )')
    plt.legend()
    plt.show()
    
roc_curve_plot(y_test, dtc.predict_proba(X_test)[:, 1].reshape(-1, 1))
```
